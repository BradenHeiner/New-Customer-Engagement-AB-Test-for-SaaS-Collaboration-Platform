###Executive Summary

After completing an exploratory analysis of newly created organizations, we identified onboarding as an area of significant opportunity to improve early collaboration and activation.

We propose a redesign of the onboarding experience aimed at reducing Time to First Collaboration for new organizations.

###What We Are Solving

Many new organizations have shown significant delays in accessing collaboration features of the platform, notably in their time to engaging with core product features and inviting their teammates. Delays in activation and collaboration may reduce long-term engagement and retention.

###Baseline Metrics

Current experience for January 2026 cohort:
- Median Time to First Collaboration Event is 530 hours
- 71% of New Organizations only invite teammates after Onboarding has begun
- % Organizations that Complete Onboarding is 38%
- p75 Onboarding Length is 2.4 hours

###Hypothesis

We believe improving the Onboarding flow will reduce Time to First Collaboration by encouraging new admins to invite teammates and engage with core froduct features sooner.

###Experiment Design

Our control will be the current onboarding process.

We will have a variant of the onboarding flow with the following changes:
- Questionnaire moved to immediately after signup
- Questionaire reduced to 3 questions, all as dropdowns: Country, Industry, and Company Size
- Questionaire required for new accounts
- Skippable Invite Your Team screen added to Onboarding, following Questionaire
- Repository Creation and Management added to New User Tour

The split between Control and Variant will be 50/50.

This test will run on all new accounts created in February 2026

###Success Threshold

Primary Metric:
- Median Time to First Collaboration (Hours)

Secondary Metrics:
- p75 Onboarding Length (Hours)
- p75 Time to First Invite (Hours)
- % Onboarding Completion

Guardrail Metric:
- % Onboarding Abandon

Variant will go live if:
- Time to First Collaboration shows material improvement (target: >10%)
- No meaningful increase in onboarding abandons

###Notes
- Required onboarding may increase experience friction for some users
- Repositories created during the new onboarding experience will be demos, and part of the tour is to delete them. These will not count as Collaboration events.
- Results may vary by organization size, particularly between solo developers and established mid-market teams.
